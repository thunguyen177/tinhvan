<b><font color="green" size = 14> The talking cat </font> </b>



Parker, “My cat! She can speak English sometimes!”

Then, he patted his cat, “Gender!”

“Meow!” the cat went.

He laughed, "Yes! That's right! But you need to practice pronunciation a little bit more! It’s male."

Mai tried to hold her laugh. 

She looked at the cat that was walking nearby and asked, “And that must be the female.”

“Yeah, sure! She’s pregnant.” 

“Okay. Suppose that she will give birth to 4 cats, is it the highest chance that she will give birth to 2 females, 2 males; 1 male, 3 females, or 3 males, 3 females?”

Parker started guessing, “Probably not 2 males, 2 females.”

“Why?”

“It seems that males and females play equal roles, but if it is 2 males, 2 females, then that’s to easy. You wouldn’t ask if that’s the case!”

“Are you sure?” Mai smiled.

Looked at Mai’s smiley face, Parker pondered for a while. He thought, “could it be…” 

“Instead of random guessing, why don’t you formulate the problem and compute the probability?” Mai said.

“Okay,” Parker nodded and started writing

“Male, male, male…”

“Why don’t you just denote 1 as the male and 0 as female?” Mai said. Her patience is low.

“Fine. The possible events are

1111;0000

1110;1101;1011;0111 (1 female, 3 females)

1100;0011;1001;0101;1010;0110 (2 males 2 females)

0001;0010;0100;1000 (3 females, 1 male)

So, the probability of 3 males, 1 female or 3 females, 1 male is 4/16+4/16=1/2. 

The probability of 2 females, 2 males, is 6/16=3/8.  

So, the probability of 3 males, 1 female and the probability of 3 females, 1 male, both are bigger than the probability of 2 males, 2 females."

“Alright, in reality, there are a lot of random variables that have values are 0 or 1 like this. For example, one can win or lose a game, the moon may or may not appear during a night, etc.” Then, we can model these events using Bernoulli distribution.

To understand this, you need to understand statistical distributions. For a discrete random variable with its corresponding event space $\Omega$, we define
$$p_X(x) =P(X=x)\forall x\in \Omega.$$
as the <font color="red"> probability mass function </font> of X. This measure the probability that X receive the value $x$. 

If you notice carefully, this notation is just a lazy way to write $P(X=x)$ by removing the $=$ sign and put $X$ to the subscript.

Obviously, 
$$0\le p_{X}(x)\le 1 \;\forall    x\in \Omega.$$ 
and Of course, the probabilities of all the events in the event space need to sum up to 1. Therefore
$$\sum _{x\in \Omega}{p_{X}}(x)=1$$
 
Come back to the cat problem. The Bernoulli distribution denoted as $Ber(p)$ is a discrete distribution that only receives value 0 or 1. Here, $p$ is the probability that the variable gets value 1, i.e., $p = P(X=1)$. Usually, $p$ is considered as the probability of success when doing something. Then, of course, the probability of a failure, which usually referred to as the probability of X, receiving value 0, is 
$$\mathbf{P} (X=0)=1-\mathbf{P} (X=1)=1-p.$$

Yet, your cat will give birth to four kitties, not one. If we consider the gender of each kitty to be independent then we can use <span style="color: red;"> (binomial distribution) </span> instead of the Bernoulli model. The binomial distribution, denoted as  <span style="color: red;">$Binom(n,p)$</span>, can help us model the possibility of $k$ success out of $n$ trials.

I mean, if you toss a coins $n$ times, independently, and the probability of getting a head (success) in each toss is $p$, then the probability of getting $k$ heads (success) is 

$$\mathbf{P} (X=k)={n \choose k}p^{k}(1-p)^{n-k}\quad {\mbox{where}}\ k=0,1,2,\dots ,n$$

and that's the distribution of a random variable $Binom(n,p)$."

"This sound like repeating the Bernoulli model $n$ times?" Parker asked.

"True. In fact, if we have $n$ independent random Bernoulli variable  $X_1,...,X_n\sim \text{Bernoulli}(p)$ then
 $$\sum_{i=1}^{n}X_i\sim Binom(n,p)$$

How many trials for the questions I asked you?"

"4"

"Ok. Then, $n = 4$. What's the probability of success?"

"$p=\frac{1}{2}$ because the probability of male and female are equal in this example" 

"Then, the probability of getting 2 males, 2 females can be computed directly by using probability distribution function:

$$P(X = 2) = {n\choose k}p^{k}(1-p)^{n-k} = {4 \choose 2}\left(\frac{1}{2} \right )^{2}(1-\frac{1}{2})^{4-2} = \frac{3}{8}$$ 

The probability of 1 male, 3 females is

$$P(X = 1) = {n \choose k}p^{k}(1-p)^{n-k} = {4 \choose 1}\left(\frac{1}{2} \right )^{1}(1-\frac{1}{2})^{4-1} = \frac{1}{4}$$

The probability of 3 male, 1 female, is

   $$P(X = 3) = {n \choose k}p^{k}(1-p)^{n-k} $$
   $$= {4 \choose 3}\left(\frac{1}{2} \right )^{3}(1-\frac{1}{2})^{4-3} =  \frac{1}{4}$$ 

"So, the probability of giving birth to at most 2 males is
 $$P(X\le 2)=P(X=0)+P(X=1)+P(X=2)=\frac{1}{16}+\frac{3}{8}+\frac{1}{4}?"$$

"Yes, but we can also compute this probability by using <font color="green"><b>Cumulative distribution function,</b></font>, which is commonly shorthanded as <font color="green"> CDF </font>. For each real number $x$, the Cumulative distribution function can be understood as the <font color="green"> probability that a random variable is smaller or equal to x </font>, i.e.,
 $$F_X(x)=P(X\leq x)$$
The Cumulative distribution function of a binomial distribution is
  $$F_X(k)=P(X\leq k)=\sum _{i=0}^k{\binom {n}{i}p^{i}(1-p)^{n-i}}$$
  where $k$ is the number of success out of $n$ trials.

"So, the probability of giving birth to at most two males is

$$F_X(2)=P(X\leq 2)=\sum _{i=0}^2{\binom {4}{i}}\left(\frac{1}{2} \right )^{i}\left(1-\frac{1}{2} \right )^{4-i}=\frac{5}{8}$$"

"But what if I want to know the probability of giving birth to 2 or 3 males?" Parker asked.

Mai replied,

"$P(2\le X\le 3)= P(X\le 3)-P(X<2)$? Am I right? Because $\{2\le X\le 3\}$ and $\{X<2\}$ are non-overlapping, i.e.   
    $$P(2\le X\le 3)+P(X<2)=P(\{2\le X\le 3\}\cup\{X<2\})=P(X\le 3)!$$

So,

$$P(2\le X\le 3)= F_X(3)-F_X(1).$$

Yet,
$$F_X(1)= P(X=0)+P(X=1)= \frac{1}{16}+\frac{3}{8}=\frac{7}{16}$$ 
$$F_X(3)=\sum _{i=0}^3{\binom {4}{i}}\left(\frac{1}{2} \right )^{i}\left(1-\frac{1}{2} \right )^{4-i}=\frac{15}{16}$$

So, the probability of giving birth to 2 or 3 males is 

$$P(2\le X\le 3)= F_X(3)-F_X(1)=\frac{15}{16}-\frac{7}{16}=\frac{8}{16}=\frac{1}{2}.$$"

"Okay, did you practice this before and trying to show off?" Parker asked.

"You gotta be kidding. My brain is not as slow as your," Mai teased him.

"Then, how do you know 

$$\sum _{i=0}^3{\binom {4}{i}}\left(\frac{1}{2} \right )^{i}\left(1-\frac{1}{2} \right )^{4-i}=\frac{15}{16}$$
that quickly?"

Mai replied,
"It's not much of calculations to do, $F_X(3)=P(X\le 3)=1-P(X>3)=1-P(X=4)=1-\frac{1}{16}=\frac{15}{16}$, okay?"

"Ok."

They kept silent for a moment, and then Parker asked again, pretending to be serious, "So, in this example, you assume that giving birth to a male is a success, and to a female is a failure. Does that applied to human also? Like me and you?"

Mai looked at him, "You mean like you and your mom? You're a success and she's a failure?"

Parker giggled, "No."
