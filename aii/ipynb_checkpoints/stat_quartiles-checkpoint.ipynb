Mai said, "Suppose I have the following data about the salary of a company (amount in dollars):


Mico:

1 000 000, 37 000 , 48 000 , 35 000,  50 000,  90 000,
  45000 , 60 000 , 42 000 , 74000 , 72 000 

Based on your ability, what do you your salary can be?"

"To be **humble**, I think I'll get an average amount. I mean the mean," Parker smiled.

Mai smiled back, "People have a tendency to evaluate themselves **higher** than they actually are. Since you say so, you probably should look at the <font color="red"> <b>first quartile </b> </font> of the data"

"Yeah, you're right! I deserve the things for the number one people!" Parker excited. 

Mai laughed, "Yeah, right! Do you know what the first quartile is about?"

Parker amazed and didn't replied.

Mai couldn't hold her laugh, "When you sort the data from low to high then the
 <font color="green"> **median** is the middle value that split the data into two halfs. One is the **lower half**, which has all the values bigger than the median and one has all the values smaller than the median. Another is the **upper half**, which comprises of all values that is bigger than the median. </font>  Next, the <font color="green"> **first quartile**, denoted by $Q_1$ , is the median of the lower half of the data set. </font> <font color="red"> This means that about 25% of the numbers in the data set lie below  $Q_1$ and about 75% lie above  $Q_1$.</font>"

"Are you teasing me? So 75% of the people will have salary higher than me and only a 25% has salary lower than me????"

"YES!" Mai loudly replied.

<p>#128554</p>

Mai continued, "AND, the <font color="green"> **third quartile**, denoted by $Q_3$ , is the median of the upper half of the data set. This means that about 75% of the numbers in the data set lie below $Q_3$ and about 25% lie above $Q_3$. </font> 

So, for this data set, sorting from low to high gives us

35 000, 37 000, 42 000, 45 000, 48 000, 50000,
 60 000, 72 000, 74 000, 90 000, 1 000 000

This data set has 11 observation. So the middle one, i.e., the one in the sixth position, which is 50000, is the median.

The lower half is 

35 000, 37 000, 42 000, 45 000, 48 000

The upper half is 

60 000, 72 000, 74 000, 90 000, 1 000 000

The first quantile is the median of the lower half. Therefore, it is 42 000."

Yet, that's when the number of observation is odd. Suppose we have the salary data where the number of observation is even:

Gogog:

1 500 000, 38 000,  60 000, 87 000, 52 000,  
     95 000, 45000, 78 000, 41 000,  74000  
     
Sorting from low to high gives

38 000, 41 000, 45 000, 52 000, 60 000,
     74 000,78 000, 87 000, 95 000,1 500 000 

Since the number of observations in this data set is even, the meadian is the mean of two middle observations: 

<p>Median = \(\frac{60 000 + 74 000}{2}= 67000\)"</p>

Mai couldn't help laughing as she looked at Parker's frowing face. She tried to make a remedy for him:

"Ok. Suppose your ability lies at the average. Then what amount should you ask for from your employer, the mean or the median?"

"The mean!"

"Ha ha... They're not gonna pay you that much!"

"Why?"

"Salaries of people such as CEO, manager, etc are variable high compared to a normal engineer. A CEO's salary can be 500 000, while a normal engineer can be 75000. So, we can call the amount of 500 000 an <font color="green">**outlier**.</font> An outlier is an observations that is so different from most of the other observations in the data set, as if it does not belong to this data set. 

In such a case, we should use median instead of the mean. If you request a mean salary, then it is $350 000, which is too high for such a position"
